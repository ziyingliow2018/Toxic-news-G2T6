```python
# import libaries

import pandas as pd
import string
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

# Import Training Data

```python
train_df = pd.read_csv("cleaned_30kundersampling.csv", sep = ",")
train_df = train_df.iloc[: , 1:]
train_df
```

# Import Test Data

```python
test_df = pd.read_csv("dataset/test_cleaned_14k.csv", sep = ",")
test_df['toxic'] = ""
test_df
```

# Test Naive Bayes on Sample data

```python
sample_test = test_df.head(100)
sample_test['toxic'] = ""
sample_test
```

```python
X_train = train_df['cleaned_text']
y_train = train_df['toxic']
X_test = sample_test['cleaned_text']
y_test = sample_test['toxic']


print("Training dataset: ", X_train.shape[0])
print("Test dataset: ", X_test.shape[0])
```

```python
count_vector = CountVectorizer(stop_words = 'english')
training_data = count_vector.fit_transform(X_train.values.astype('U'))
testing_data = count_vector.transform(X_test.values.astype('U'))
```

```python
naive_bayes = MultinomialNB()
naive_bayes.fit(training_data, y_train)
```

```python
predictions_test = naive_bayes.predict(testing_data)
predictions_test
```

```python
# Add predictions to dataframe
predictions_df_sample = sample_test

predictions_df_sample['Predicted'] = predictions_test.tolist()
predictions_df_sample.drop(columns=['toxic'],inplace=True)
predictions_df_sample

# Write to csv
# prediction_df_bagging.to_csv('Predictions/BaggingPredictions.csv',index=False)
# prediction_df_bagging
```

# Naive Bayes 14k data

```python
X_train = train_df['cleaned_text']
y_train = train_df['toxic']
X_test = test_df['cleaned_text']
y_test = test_df['toxic']


print("Training dataset: ", X_train.shape[0])
print("Test dataset: ", X_test.shape[0])
```

```python
count_vector = CountVectorizer(stop_words = 'english')
training_data = count_vector.fit_transform(X_train.values.astype('U'))
testing_data = count_vector.transform(X_test.values.astype('U'))
```

```python
naive_bayes = MultinomialNB()
naive_bayes.fit(training_data, y_train)
```

```python
predictions = naive_bayes.predict(testing_data)
predictions
```

```python
# Add predictions to dataframe
prediction_df_14k = test_df

prediction_df_14k['Predicted'] = predictions.tolist()
prediction_df_14k.drop(columns=['toxic'],inplace=True)
prediction_df_14k
```

```python
# Write to csv
prediction_df_14k.to_csv('naive_bayes_predictions_14k.csv',index=False)
```

# Model Performance

```python
from sklearn.metrics import confusion_matrix
```

```python
# Get the confusion matrix
conf_matrix = confusion_matrix(y_test, predictions)

# Retrieve TN, FN, TP and FP
tn = conf_matrix[1][1]
fn = conf_matrix[0][1]
tp = conf_matrix[0][0]
fp = conf_matrix[1][0]

accuracy =  (tp + tn)/(tp + tn + fp + fn)
precision = tp/(tp+fp)
recall = tp/(tp+fn)
f1_score = 2* ((recall * precision) / (precision+recall))
```
